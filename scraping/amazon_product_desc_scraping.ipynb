### Need to pass header information when trying to access amazon url

```python
def getAmazonPage(productUrl):
    headers = {
        'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10.11; rv:10.0) Gecko/20100101 Firefox/10.0',
    }
    res = requests.get(productUrl, headers=headers)
    res.raise_for_status()
    
#     page = response.text
#     soup = BeautifulSoup(page)

    soup = BeautifulSoup(res.text, 'html.parser')
#     elems = soup.select('#newOfferAccordionRow .header-price')
    return soup
```

```python
page_soup = getAmazonPage('https://www.amazon.com/dp/B00IZ9R6N2')
```
